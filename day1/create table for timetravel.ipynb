```
%python
# Create database if not exists
spark.sql("CREATE DATABASE IF NOT EXISTS car_data_delta")

# Read CSV
df = spark.read.csv("/FileStore/tables/car_data-3.csv", header=True, inferSchema=True)

# Save as Delta table in the database
df.write.format("delta").mode("overwrite").saveAsTable("car_data_delta")
```
